```python
from dotenv import load_dotenv
import os
load_dotenv()
groq_api_key=os.getenv("GROQ_API_KEY")

if groq_api_key:
    print("GROQ API KEY exists in the env")
else:
    print("GROQ API KEY does not exists")
```

```python
groq_base_url="https://api.groq.com/openai/v1"
```

```python
from openai import OpenAI

groq=OpenAI(
    api_key=groq_api_key,
    base_url= groq_base_url
)
```

```python
tell_a_joke=[
    {"role":"user",
     "content":"Tell me a joke on AI"}
]
```

```python
response=groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=tell_a_joke
)
print(response.choices[0].message.content)
```

```python
easy_puzzle=[
    {"role":"user",
     "content":"2 coins are tossed and one them turns out to be head what is the probability that other one turns out to be tail? Give only probailtity"
     }
]

response=groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=easy_puzzle,
    reasoning_effort="low"
)
print(response.choices[0].message.content)
```

```python
from IPython.display import display, Markdown
display(Markdown(response.choices[0].message.content))
```

```python
response=groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=easy_puzzle,
    reasoning_effort="medium"
)
display(Markdown(response.choices[0].message.content))
```

```python
response=groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=easy_puzzle,
    reasoning_effort="high"
)
display(Markdown(response.choices[0].message.content))
```

```python
dilemma_prompt=""" You and a patner are contestants on a game show. You're each taken to separate rooms and given a choice:
1. Cooperate:Choose "share"- if both of you choose this, you each win $1000.
2. Defect:Choose "steal"- if one stands and the other shares, the stealer gets $2000 and the sharer gets nothing.
If you both steal, you both get nothing.
Do you choose to Steal or Share? Pick one. 
"""

dilemma=[{
    "role":"user",
    "content":dilemma_prompt
}]


```

```python
response=groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=dilemma
)

display(Markdown(response.choices[0].message.content))
```

```python
response=groq.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=dilemma
)

display(Markdown(response.choices[0].message.content))
```

```python
response=groq.chat.completions.create(
    model="qwen/qwen3-32b",
    messages=dilemma
)

display(Markdown(response.choices[0].message.content))
```
