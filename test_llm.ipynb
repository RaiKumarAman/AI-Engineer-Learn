```python
from dotenv import load_dotenv
import os
load_dotenv()
groq_api_key=os.getenv("GROQ_API_KEY")

if groq_api_key:
    print("GROQ API KEY exists in the env")
else:
    print("GROQ API KEY does not exists")
```

```python
groq_base_url="https://api.groq.com/openai/v1"
```

```python
from openai import OpenAI

groq=OpenAI(
    api_key=groq_api_key,
    base_url= groq_base_url
)
```

```python
tell_a_joke=[
    {"role":"user",
     "content":"Tell me a joke on AI"}
]
```

```python
response=groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=tell_a_joke
)
print(response.choices[0].message.content)
```

```python
gpt_system="You are a financial consultant in India who has a experience of about 20+ years. You are sitting in fornt of a customer who is about to start his investement journey and will ask you question about it. Now your goal is to help him make the right decision. Reply as a human would talk"

llama_system="You are a newbie in India going to start you investement journey and is sitting in a room with financial expert ask him questions as you like and until you are satisfied. Reply as if conversation going on not the paragraphs "
```

```python
gpt_message=["Hi"]
llama_message=["Hey, Thank you for appointment."]
```

```python
def call_gpt():
    messages=[{"role":"system","content":gpt_system}]
    for gpt, llama in zip(gpt_message, llama_message):
        messages.append({"role":"assistant", "content":gpt})
        messages.append({"role":"user", "content":llama})
    
    response=completion(
    model="groq/openai/gpt-oss-20b",
    api_key=groq_api_key,
    messages=messages
    )

    return response.choices[0].message.content

```

```python
def call_llama():
    messages=[{"role":"system","content":gpt_system}]
    for gpt, llama in zip(gpt_message, llama_message):
        messages.append({"role":"assistant", "content":llama})
        messages.append({"role":"user", "content":gpt})
    
    response=completion(
    model="groq/meta-llama/llama-4-scout-17b-16e-instruct",
    api_key=groq_api_key,
    messages=messages
    )

    return response.choices[0].message.content
```

```python
i=0
while(i<5):
    gpt_next=call_gpt()
    display(Markdown(f'###GPT: {gpt_next}\n'))
    gpt_message.append(gpt_next)

    llama_next=call_llama()
    display(Markdown(f'###LLAMA: {llama_next}'))
    llama_message.append(llama_next)
    i+=1
```
