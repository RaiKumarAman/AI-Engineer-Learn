```python
import gradio as gr
```

```python
groq_api_key=os.getenv("GROQ_API_KEY")

if groq_api_key:
    print("GROQ API KEY exists in the env")
else:
    print("GROQ API KEY does not exists")
```

```python
groq_base_url="https://api.groq.com/openai/v1"
```

```python
llm=OpenAI(api_key=groq_api_key, base_url=groq_base_url)
```

```python
system_message="You are a helpful assisant who has knowledge in investment and financial planning. And your goal is to help people make good investement decisions."

def gpt(prompts):
    messages=[{"role":"system", "content":system_message},
              {"role":"user","content":prompts}]
    response=llm.chat.completions.create(model="openai/gpt-oss-120b",
                                         messages=messages)
    return response.choices[0].message.content
```

```python
system_message="You are a helpful assisant."

def gpt_stream(prompts):
    messages=[{"role":"system", "content":system_message},
              {"role":"user","content":prompts}]
    stream=llm.chat.completions.create(model="openai/gpt-oss-120b",messages=messages,stream=True
                                        )
    result=""
    for i in stream:
        result+=i.choices[0].delta.content or ""
    yield result
```

```python
message_input=gr.Textbox(label="Your text", info="Enter a message for GPT", lines=7)
message_output=gr.Markdown(label="Response:")
gr.Interface(fn=gpt_stream,
    title="GPT",
    inputs=[message_input], outputs=[message_output], flagging_mode="never").launch(share=True)
```
